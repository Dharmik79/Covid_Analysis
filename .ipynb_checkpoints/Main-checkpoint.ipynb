```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import plotly
import plotly.express  as px # wrapper of plotly
import plotly.graph_objects as go
style.use('ggplot')
import cufflinks as cf
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot
import folium
plt.rcParams['figure.figsize']=(15,8)
```

```python
# connect plotly to pujyter notebook
pyo.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df=pd.read_excel('Covid cases in India.xlsx',index_col=0)
df.head()
```

```python
df['Total_Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
df['Active_Cases']=df['Total_Cases']-(df['Cured']+df['Death'])
df.head()
```

```python
df.style.background_gradient(cmap="Reds")   # df style to find the highest number 
```

```python
# group by takes 2 columns for the vaalues
Total_Active_Cases=df.groupby('Name of State / UT')['Active_Cases'].sum().sort_values(ascending=False)
```

```python
# to frame convert to dataframe
Total_Active_Cases.to_frame().style.background_gradient(cmap="Reds")
```

```python
# Graphical Representation

df.plot(kind="bar",x='Name of State / UT',y='Total_Cases')
```

```python
# same above figure in plotly

df.iplot(kind="bar",x='Name of State / UT',y='Total_Cases')
```

```python
px.bar(df,x='Name of State / UT',y='Total_Cases')
```

```python
Indian_Coord=pd.read_excel('Indian Coordinates.xlsx')
Indian_Coord
```

```python
df_full=pd.merge(Indian_Coord,df,on='Name of State / UT')
df_full
```

```python
# Make a Map
map=folium.Map(location=[40,70],zoom_start=6,tiles='Stamenterrain')

for lat,long,value,name in zip(df_full['Latitude'],df_full['Longitude'],df_full['Total_Cases'],df_full['Name of State / UT']):
    folium.CircleMarker([lat,long],radius=value*0.4,popup=('<strong>State</strong>: '+str(name).capitalize()+'<br><strong>Total Cases<strong>:'+str(value)+'<br>'),color='red',fill_color='red',fill_opacity=0.3).add_to(map)
```

```python
df=pd.read_csv('world.csv',parse_dates=True,index_col=0)
```

```python
df.head()
```

```python
df.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
confirmed=df.groupby('Date').sum()['Confirmed'].reset_index()
death=df.groupby('Date').sum()['Deaths'].reset_index()
rec=df.groupby('Date').sum()['Recovered'].reset_index()
```

```python

fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))

fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=rec['Date'],y=rec['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```

```python
df_confirmed=pd.read_csv('confirmed.csv')
```

```python
df_confirmed.rename(columns={'Country/Region':'Country'},inplace=True)
```

```python
df_latlong=pd.merge(df,df_confirmed,on=['Country','Province/State'])
```

```python
fig=px.density_mapbox(df_latlong,lat="Lat",lon="Long",hover_name="Province/State",hover_data=["Confirmed","Deaths","Recovered"],animation_frame="Date",color_continuous_scale="Portland",radius=7,zoom=0,height=700)
fig.update_layout(title='Worldwide Corona Virus Cases')
fig.update_layout(mapbox_style="open-street-map",mapbox_center_lon=0)

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
```
